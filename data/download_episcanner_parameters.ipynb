```python
import requests
import numpy as np
import pandas as pd
from itertools import product
import matplotlib.pyplot as plt
```

```python
code_to_state = {33: 'RJ', 32: 'ES', 41: 'PR', 23: 'CE', 21: 'MA',
 31: 'MG', 42: 'SC', 26: 'PE', 25: 'PB', 24: 'RN', 22: 'PI', 27: 'AL',
 28: 'SE', 35: 'SP', 43: 'RS', 15: 'PA', 16: 'AP', 14: 'RR',  11: 'RO',
 13: 'AM', 12: 'AC', 51: 'MT', 50: 'MS', 52: 'GO', 17: 'TO', 53: 'DF',
 29: 'BA'}

code_to_state.values()
```

```python
%%time
episcanner_api = "https://api.mosqlimate.org/api/datastore/episcanner/"

for state in code_to_state.values():
    df = pd.DataFrame()

    for year in  np.arange(2011, 2025):
        params = {
            "disease": "dengue",
            "uf": f"{state}",
            "year": f"{year}",
        }

        try:
            resp = requests.get(episcanner_api, params=params)
            
            items = resp.json()
            df_ = pd.DataFrame(items)
        except:
            print(state)
            print(year)

        df = pd.concat([df, df_])

    df.to_csv(f'episcanner_{state}.csv.gz')
    
```
